```python
#Tokenizar el texto  
documento_spacy = nlp(documento)
```

```python
documento_spacy
```

```python
for token in documento_spacy:
#Trato de determinar que palabras son StopWords o Signos de puntuación 
    print(token.text,'-->',token.is_stop,'-->',token.is_punct)
```

```python
#Cre una lista con las plabras que me interesan usando Comprehension lists
documento_procesado = [token for token in documento_spacy if not token.is_stop and not token.is_punct]
```

```python
documento_procesado
```

```python
# Lematización : Lleva todo a su palabra raiz. 
for token in documento_procesado:
    print(token.lemma_)
```

```python
# Imprimir emails
for token in documento_procesado:
    if token.like_email:
        print(token.text)
```

```python
# Intento capturar montos:
for token in documento_procesado:
    if token.like_num:
        print(token.text)
```

```python
#Creo que esto es maás interesante para nuestro proyecto las entidades. 
print(documento_spacy.ents)
```

```python
# Ver NER
for entidad in documento_spacy.ents:
    print(entidad.text, '--', entidad.label_)
```

```python
# En el anterior parece que Reconoce Localización bien en las entidades.
# Veamos lo anterior visualmente. 

from spacy import displacy
displacy.render(documento_spacy,style = 'ent', jupyter=True)
```
